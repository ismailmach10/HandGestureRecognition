```python
import cv2
import os

BASE_PATH = "C:/Users/ismai/HandGestureRecognition/data/webcam_gestures"

def capture_gesture(gesture_name, num_images=50):
    save_path = os.path.join(BASE_PATH, gesture_name)
    os.makedirs(save_path, exist_ok=True)

    cap = cv2.VideoCapture(0)
    count = 0

    print(f"=== Capture du geste : {gesture_name} ===")
    print("ðŸ‘‰ Mets ta main dans le carrÃ© vert")
    print("ðŸ‘‰ Appuie sur ESPACE pour prendre une photo")
    print("ðŸ‘‰ Appuie sur Q pour quitter")

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        frame = cv2.flip(frame, 1)

        # CADRE VERT
        x1, y1, x2, y2 = 100, 100, 400, 400
        roi = frame[y1:y2, x1:x2]

        cv2.rectangle(frame, (x1, y1), (x2, y2), (0,255,0), 2)

        cv2.imshow("Webcam - Mets ta main dans le cadre", frame)

        key = cv2.waitKey(1)

        if key == 32:  # ESPACE
            gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
            gray = cv2.resize(gray, (64, 64))
            cv2.imwrite(f"{save_path}/{count}.jpg", gray)
            count += 1
            print(f"Photo {count}/{num_images}")

        if key == ord('q') or count >= num_images:
            break

    cap.release()
    cv2.destroyAllWindows()
    print("âœ… Capture terminÃ©e\n")
```

```python
import cv2
import os
import time

# --- CONFIGURATION ---
# Liste de tes gestes
gestures = ["fist", "palm", "victory"]
print("Quel geste voulez-vous enregistrer ?")
for i, g in enumerate(gestures):
    print(f"{i} : {g}")

choice = int(input("Tapez le nombre (0, 1 ou 2) : "))
gesture = gestures[choice]

save_dir = f"C:/Users/ismai/HandGestureRecognition/data/webcam_gestures/{gesture}"
os.makedirs(save_dir, exist_ok=True)

cap = cv2.VideoCapture(0)
count = 0
max_images = 300
recording = False

print(f"\nPrÃªt pour : {gesture.upper()}")
print("Instructions :")
print("1. Placez votre main dans le cadre.")
print("2. Appuyez sur 'S' pour dÃ©marrer.")
print("3. BOUGEZ votre main (rotation, avant/arriÃ¨re) pendant la capture !")

while count < max_images:
    ret, frame = cap.read()
    if not ret: break
    
    frame = cv2.flip(frame, 1)
    h, w, _ = frame.shape
    size = 300
    x1, y1 = (w - size) // 2, (h - size) // 2
    
    # Zone d'intÃ©rÃªt
    roi = frame[y1:y1+size, x1:x1+size]
    
    # --- TRAITEMENT ROBUSTE ---
    gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5, 5), 2)
    # Le seuillage adaptatif aide Ã  sÃ©parer la main du fond
    _, thresholded = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    
    img_final = cv2.resize(thresholded, (64, 64))

    # --- UI ---
    cv2.rectangle(frame, (x1, y1), (x1+size, y1+size), (0, 255, 0), 2)
    color = (0, 0, 255) if recording else (0, 255, 0)
    status = "RECORDING..." if recording else "WAITING (Press 'S')"
    
    cv2.putText(frame, f"Geste: {gesture} | {count}/{max_images}", (20, 40), 
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)
    cv2.putText(frame, status, (20, 70), cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

    cv2.imshow("Main Camera", frame)
    cv2.imshow("AperÃ§u ModÃ¨le (Noir et Blanc)", img_final)

    key = cv2.waitKey(1)
    if key == ord('s'):
        recording = True
    
    if recording:
        # On enregistre l'image en Noir et Blanc (beaucoup plus prÃ©cis pour le modÃ¨le)
        cv2.imwrite(f"{save_dir}/{gesture}_{count}.png", img_final)
        count += 1
        # Petit dÃ©lai pour avoir le temps de bouger la main
        time.sleep(0.02) 

    if key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
print(f"\nSuccÃ¨s ! 300 images de '{gesture}' sont dans {save_dir}")
```
