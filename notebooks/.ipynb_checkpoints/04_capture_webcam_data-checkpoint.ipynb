```python
import cv2
import os

BASE_PATH = "C:/Users/ismai/HandGestureRecognition/data/webcam_gestures"

def capture_gesture(gesture_name, num_images=50):
    save_path = os.path.join(BASE_PATH, gesture_name)
    os.makedirs(save_path, exist_ok=True)

    cap = cv2.VideoCapture(0)
    count = 0

    print(f"=== Capture du geste : {gesture_name} ===")
    print("ðŸ‘‰ Mets ta main dans le carrÃ© vert")
    print("ðŸ‘‰ Appuie sur ESPACE pour prendre une photo")
    print("ðŸ‘‰ Appuie sur Q pour quitter")

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        frame = cv2.flip(frame, 1)

        # CADRE VERT
        x1, y1, x2, y2 = 100, 100, 400, 400
        roi = frame[y1:y2, x1:x2]

        cv2.rectangle(frame, (x1, y1), (x2, y2), (0,255,0), 2)

        cv2.imshow("Webcam - Mets ta main dans le cadre", frame)

        key = cv2.waitKey(1)

        if key == 32:  # ESPACE
            gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
            gray = cv2.resize(gray, (64, 64))
            cv2.imwrite(f"{save_path}/{count}.jpg", gray)
            count += 1
            print(f"Photo {count}/{num_images}")

        if key == ord('q') or count >= num_images:
            break

    cap.release()
    cv2.destroyAllWindows()
    print("âœ… Capture terminÃ©e\n")
```

```python
import cv2
import os

gesture = "fist"  # CHANGE ICI: fist / palm / victory
save_dir = f"C:/Users/ismai/HandGestureRecognition/data/webcam_gestures/{gesture}"
os.makedirs(save_dir, exist_ok=True)

cap = cv2.VideoCapture(0)

count = 0
max_images = 300

while True:
    ret, frame = cap.read()
    if not ret:
        break

    h, w, _ = frame.shape
    size = 300
    cx, cy = w // 2, h // 2

    roi = frame[
        cy - size//2 : cy + size//2,
        cx - size//2 : cx + size//2
    ]

    cv2.rectangle(
        frame,
        (cx - size//2, cy - size//2),
        (cx + size//2, cy + size//2),
        (0, 255, 0),
        2
    )

    cv2.putText(
        frame,
        f"{gesture} | Images: {count}/{max_images}",
        (30, 40),
        cv2.FONT_HERSHEY_SIMPLEX,
        1,
        (0, 255, 0),
        2
    )

    cv2.imshow("Capture Gesture", frame)

    key = cv2.waitKey(1)

    if key == ord("s") and count < max_images:
        img = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
        img = cv2.resize(img, (64, 64))
        cv2.imwrite(f"{save_dir}/{gesture}_{count}.png", img)
        count += 1

    if key == ord("q") or count >= max_images:
        break

cap.release()
cv2.destroyAllWindows()
```
